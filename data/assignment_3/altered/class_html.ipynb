```python
import os
import pandas as pd
import sqlite3
from bs4 import BeautifulSoup
```

```python
data_dir = '/Users/bradysengkhounmany/data_wrangling_304/data/assignment_3/altered/downloaded_class_submissions'
html_files = [f for f in os.listdir(data_dir) if f.endswith(".html")]
```

```python
all_rows = []
for file_name in html_files:
    file_path = os.path.join(data_dir, file_name)

    with open(file_path, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    h1_tag = soup.find("h1")
    if h1_tag:
        name_value = h1_tag.text.strip()

    table = soup.find("table")
    if table:
        headers = [th.text.strip() for th in table.find_all("th")]
        headers.append('Name')


        for row in table.find_all('tr')[1:]:
            cols = [td.text.strip() for td in row.find_all('td')]
            cols.append(name_value)
            all_rows.append(cols)

favorites = pd.DataFrame(all_rows, columns=headers)


favorites.to_csv('favorites.csv', index=False, encoding='utf-8')
favorites
```
