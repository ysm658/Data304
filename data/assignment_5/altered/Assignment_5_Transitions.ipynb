```python
import pandas as pd
```

```python
paths = pd.read_csv("course_pathways.csv")
catalog = pd.read_csv("course_catalog_by_major.csv")
```

```python
majors_cecs = ["Data Science", "Artificial Intelligence", "Cybersecurity"]
filtered_df = paths[paths["Major"].isin(majors_cecs)]
filtered_df
```

```python
required = catalog[catalog["Type"] == "Required"]
transitions = []

for _, row in filtered_df.iterrows():
    major = row["Major"]
    student_id = row["StudentID"]
    courses = [c.strip() for c in row["CoursePath"].split("->")]
    required_courses = required[required["Major"] == major]["Course"].tolist()
    required_only = [c for c in courses if c in required_courses]
    
    for i in range(len(required_only) - 1):
        current = required_only[i]
        future = required_only[i + 1]
        transitions.append((current, future))

transitions_df = pd.DataFrame(transitions, columns=["current", "future"])
transitions_df = transitions_df.value_counts().reset_index(name="count")
transitions_df
```

```python
transitions_df.to_csv("transitions.csv", index=False)
```
